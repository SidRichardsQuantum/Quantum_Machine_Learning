# Variational Quantum Regressor

Minimal package-client example for the PennyLane-based regression workflow in `qml.regression`.

```python
from qml.regression import run_vqr
```

## Run a minimal variational regression experiment

```python
result = run_vqr(
    n_samples=100,
    noise=0.1,
    test_size=0.25,
    seed=0,
    n_layers=2,
    steps=50,
    step_size=0.1,
    plot=True,
    save=True,
)
```

## Summary metrics

```python
print("Train MSE:", result["train_mse"])
print("Test MSE:", result["test_mse"])
print("Train MAE:", result["train_mae"])
print("Test MAE:", result["test_mae"])
print("Final loss:", result["final_loss"])
```
